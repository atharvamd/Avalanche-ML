```python
# convert numerical column to a categorical datatype
df['lulcdt'] = pd.Categorical(df['lulcdt'])

# perform one-hot encoding on the categorical column
one_hot = pd.get_dummies(df['lulcdt'], prefix='lulcdt')

# concatenate the original dataframe with the one-hot encoded dataframe
df = pd.concat([df, one_hot], axis=1)

# drop the original categorical column
df = df.drop('lulcdt', axis=1)
```

```python
new_order = ['elev','asp','chnlndst','pln','prfl','slp','tp','tr','tw','lulcdt_1.0','lulcdt_2.0','lulcdt_3.0','lulcdt_4.0','lulcdt_5.0','lulcdt_6.0','target']
df = df[new_order]
```

```python
df
```

```python
#Making the dataset into 1:1 ratio of the avalanche and non-avalanche
#Split the original DataFrame into two DataFrames based on the class labels
df_majority = df[df['target'] == 0]
df_minority = df[df['target'] == 1]
```

```python
#Combine the minority class DataFrame with the downsampled majority class DataFrame
df_majority_downsampled = df_majority.sample(n=df_minority.shape[0], random_state=42)
df_balanced = pd.concat([df_majority_downsampled, df_minority])
```

```python
#Shuffle the rows of the resulting balanced DataFrame to randomize the order of the samples
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df_balanced
```

```python
df_balanced.describe()
```

```python
y = df_balanced.target
```

```python
features = ['elev','asp','chnlndst','pln','prfl','slp','tp','tr','tw','lulcdt_1.0','lulcdt_2.0','lulcdt_3.0','lulcdt_4.0','lulcdt_5.0','lulcdt_6.0']
X = df_balanced[features]
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y,test_size=0.3,random_state = 0)
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
#These were the best parameters for the Random Forest model
forest_model = RandomForestClassifier(n_estimators=250,criterion='entropy',max_depth=60,max_features=30,class_weight='balanced_subsample',oob_score=True,bootstrap=True,random_state=1)
forest_model.fit(train_X, train_y)
preds = forest_model.predict(test_X)

from sklearn.metrics import accuracy_score
print('accuracy:',accuracy_score(test_y, preds))

from sklearn.metrics import roc_auc_score
print('ROC score:',roc_auc_score(test_y, preds))
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=100, solver='lbfgs', class_weight='balanced')

# Train the classifier on the training set
lr.fit(train_X, train_y)

# Make predictions on the test set
preds = lr.predict(test_X)


print('accuracy:',accuracy_score(test_y, preds))


print('ROC score:',roc_auc_score(test_y, preds))
```

```python
#Support Vector Machine(SVM)
from sklearn.svm import SVC

# Create an SVM classifier with specified parameters
svm = SVC(kernel='rbf', C=100,gamma='scale',class_weight='balanced')

# Train the classifier on the training set
svm.fit(train_X, train_y)

# Make predictions on the test set
preds = svm.predict(test_X)


print('accuracy:',accuracy_score(test_y, preds))


print('ROC score:',roc_auc_score(test_y, preds))
```

```python
#KNN Classifier
from sklearn.neighbors import KNeighborsClassifier

# Create KNN classifier with k = 3
knn = KNeighborsClassifier(n_neighbors=3)

# Fit the classifier to the training data
knn.fit(train_X, train_y)

# Predict the classes of the testing set
preds = knn.predict(test_X)


print('accuracy:',accuracy_score(test_y, preds))


print('ROC score:',roc_auc_score(test_y, preds))
```

```python
#XGBoost Classifier
import xgboost as xgb

model = xgb.XGBClassifier(learning_rate=0.01, max_depth=5, n_estimators=100)

# Train the model
model.fit(train_X, train_y)

# Evaluate the model
preds = model.predict(test_X)


print('accuracy:',accuracy_score(test_y, preds))


print('ROC score:',roc_auc_score(test_y, preds))
```
